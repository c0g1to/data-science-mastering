```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
%matplotlib inline

plt.style.use('default')
```

```python
lang = pd.read_csv('files/w3bar1.txt', sep="[,:]\ ", index_col=0, header=None, engine='python').T
fig, ax = plt.subplots()
ax.pie(lang.Popularity, labels=lang['Programming languages'], autopct='%.1f%%',
       startangle=135, explode=[0.1, 0, 0, 0, 0, 0], shadow=True,
       wedgeprops={'edgecolor': 'k'})
ax.axis('equal')
```

```python
ax.set_title("Popularity of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago",
#             backgroundcolor='lightgray',
             bbox={'facecolor':'0.8', 'edgecolor':'k', 'pad':5}, pad=16)
fig
```
