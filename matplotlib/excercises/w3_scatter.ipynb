```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd 
import numpy as np
import io
%matplotlib inline

plt.style.use('default')
```

```python
plt.scatter(np.random.randn(200), np.random.randn(200), s=3, c='r')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.scatter(np.random.randn(50), np.random.randn(50), s=50, facecolor='none', edgecolor='g', marker='o')
```

```python
colors = np.random.random((5, 3))
colors[np.random.randint(0, colors.shape[0], 50)]
plt.scatter(np.random.randn(50), np.random.randn(50),
            s=np.random.randint(20, 200, 50), marker='o',
            facecolor=colors[np.random.randint(0, colors.shape[0], 50)], edgecolor='k', alpha=0.5)
```

```python
data = [row.replace(']', '').replace(' = [', ',') for row in open('files/w3scatter4.txt')]
test_df = pd.read_csv(io.StringIO(''.join(data)), index_col=0, header=None, skiprows=1)

for subject, color in zip(test_df.index[:2], ['red', 'green']):
    plt.scatter(test_df.loc['marks_range'], test_df.loc[subject], label=subject, s=10, c=color)
plt.legend()
```

```python
widths = np.random.uniform(0, 100, (30, 3))
heights = np.random.uniform(0, 300, (30, 3))
colors = np.random.random((3, 3))
for i in range(widths.shape[1]):
    plt.scatter(widths[:, i], heights[:, i], color=colors[i], s=10)
plt.xlabel('weight')
plt.ylabel('height')
plt.title('Groupwise Weight vs Height scatter plot')
```
