```
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import IPython
%matplotlib inline

plt.style.use('default')
```

```
sales = pd.read_csv('files/company_sales_data.csv')
sales
```

```
plt.plot(sales.month_number, sales.total_profit)
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.yticks(np.arange(1, 5+1, 1) * 100000)
plt.title('Company profit per month')
# plt.axis([0, 13, 100000, 500000])
```

```
plt.plot(sales.month_number, sales.total_profit, 'r--o', markerfacecolor='k', lw=3,
    label='Profit data of last year')
plt.legend(loc='lower right')
plt.yticks(np.arange(1, 5+1, 1) * 100000)
```

```
for col in sales.columns[1:-2]:
    plt.plot(sales.month_number, sales[col], label=col, lw=3, marker='o')
plt.legend(loc='best')
plt.yticks(np.array([1, 2, 4, 6, 8, 10, 12, 15, 18]) * 1000)
```

```
plt.scatter(sales.month_number, sales.toothpaste, label='Tooth paste Sales data')
plt.grid(True, ls='--', lw=1)
plt.xticks(np.arange(1, 12+1, 1))
plt.legend()
```

```
w = 0.4
i = -1
for col in sales.columns[1:3]:
    plt.bar(sales.month_number + w * i, sales[col], label=col, width=-w * i, align='edge')
    i *= -1
plt.grid(True, ls='--', lw=1)
plt.xticks(np.arange(1, 12+1, 1))
plt.legend()
```

```
plt.bar(sales.month_number, sales.bathingsoap)
plt.gcf().savefig('files/bathingsoap.png', dpi=400)
```

```
intervals = (np.array([150, 175, 200, 225, 250, 300, 350]) * 1000)
plt.hist(sales.iloc[:, -1], bins=intervals, label='Profit data')
plt.xticks(intervals)
plt.legend()
```

```
plt.pie(sales.iloc[:, 1:-2].sum(), labels=sales.columns[1:-2], autopct='%.1f%%')
plt.legend(loc='lower right')
plt.axis('equal')
```

```
fig, ax = plt.subplots(2, sharex=True)
ax[0].plot(sales.month_number, sales.bathingsoap, 'ko-', lw=3)
ax[1].plot(sales.month_number, sales.facewash, 'ro-', lw=3)
ax[1].set_yticks([1500, 2000])
ax[0].set_yticks([7500, 10000, 12500])
ax[1].set_xlabel('Month number')
ax[1].set_ylabel('Sales units in number')
ax[1].set_title('Sales data of a facewash')
ax[0].set_title('Sales data of a Bathingsoap')
ax[1].set_xticks(sales.month_number)
```
