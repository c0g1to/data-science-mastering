```python
X = [10, 20, 30]
Y1 = [20, 40, 10]
Y2 = [40, 10, 30]
plt.plot(X, Y1, label='line 1')
plt.plot(X, Y2, label='line 2')
plt.legend()
```

```python
plt.plot(X, Y1, label='line1-width-3', color='blue', linewidth=3)
plt.plot(X, Y2, label='line2-width-5', color='red', linewidth=5)
plt.legend()
```

```python
finance = pd.read_csv('files/w3b4.txt', index_col=0, parse_dates=True)
fig, ax = plt.subplots()

# ax.set_xticks(finance.index[::1])
# ax.set_xticklabels(
#     [date.strftime("%Y-%m-%d") for date in finance.index])
#or
ax.xaxis.set_major_locator(mpl.dates.DayLocator())
ax.xaxis.set_major_formatter(mpl.dates.DateFormatter('%Y-%m-%d'))

ax.plot(finance.index, finance.Close, '-or')
```

```python
ax.grid(True, linewidth=.5, color='blue')
fig
```

```python
ax.grid(True, lw=.5, c='r', which='major')
ax.grid(True, lw=.3, ls='--', c='gray', which='minor')

# ax.xaxis.set_minor_locator(mpl.ticker.AutoMinorLocator())
# ax.yaxis.set_minor_locator(mpl.ticker.AutoMinorLocator())
#or 
ax.minorticks_on()

ax.tick_params(which='both', bottom=False, left=False)
fig
```

```python
axes = []
grid = plt.GridSpec(2, 3)
axes.append(plt.subplot(grid[0, :]))
for i in range(0, 3):
    axes.append(plt.subplot(grid[1, i]))
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])
```
