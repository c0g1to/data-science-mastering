```python
#@formatter:off
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
%matplotlib inline
#@formatter:on
```

```python
x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
```

```python
ax1.plot(x, y)
ax2.plot(x, y)
fig
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
ax1.plot(x, y)
ax2.plot(y, z)
fig
```

```python
fig, axes = plt.subplots(1, 2)
```

```python
axes[0].plot(x, y)
axes[1].plot(x, z)
fig
```
