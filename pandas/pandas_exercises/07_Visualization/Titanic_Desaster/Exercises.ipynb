### Step 5. Create a pie chart presenting the male/female proportion

```python
titanic['male'] = titanic.sex == 'male'
# titanic.drop(columns='sex', inplace=True)
```

```python
males = titanic.male.sum()
females = titanic.shape[0] - males
plt.title('Sex Proportion')
plt.pie([males, females], labels=['Males', 'Females'], autopct='%1.1f%%',
        explode=[0.0, 0.1], startangle=90, colors='br')
plt.tight_layout()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
fig, ax = plt.subplots()
ax.set(title='Fare x Age', xlabel='Age', ylabel='Fare')
scatter = ax.scatter(x=titanic.age, y=titanic.fare, c=titanic.male)

handles, labels = scatter.legend_elements()
legend = ax.legend(handles=handles[:2], labels=labels[:2], title='Male?')
ax.add_artist(legend)
```

```python
#or
sns.scatterplot(data=titanic, x='age', y='fare', hue='sex', palette='Set1')
```
